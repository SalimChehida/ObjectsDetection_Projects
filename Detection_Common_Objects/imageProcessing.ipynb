```python
def load_and_print_image(url):
    try:
        # Send a GET request to fetch the image
        response = requests.get(url)

        # Check if the request was successful
        response.raise_for_status()

        # Read the image data
        image_data = BytesIO(response.content)

        # Open the image using PIL (Pillow)
        image = Image.open(image_data)

        # Display the image
        image.show()

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred: {err}")
```

```python
from IPython.display import display
from PIL import Image
import requests
from io import BytesIO

def show_image_from_web(url):
    try:
        # Send a GET request to fetch the image
        response = requests.get(url)

        # Check if the request was successful
        response.raise_for_status()

        # Read the image data
        image_data = BytesIO(response.content)

        # Open the image using PIL (Pillow)
        image = Image.open(image_data)

        # Display the image in the Jupyter Notebook
        display(image)

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred: {err}")

```

```python
def load_and_save_image(url, save_path):
    try:
        # Send a GET request to fetch the image
        response = requests.get(url)

        # Check if the request was successful
        response.raise_for_status()

        # Read the image data
        image_data = BytesIO(response.content)

        # Open the image using PIL (Pillow)
        image = Image.open(image_data)

        # Save the image to the specified path
        image.save(save_path)

        print(f"Image saved successfully to {save_path}")

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An error occurred: {err}")
```

```python
if __name__ == "__main__":
    # Replace 'YOUR_IMAGE_URL' with the URL of the image you want to load and save
    image_url = 'https://ultralytics.com/images/bus.jpg'
    image_url1 = 'https://www.freecodecamp.org/news/content/images/2023/04/cat_dog.jpg'

    # Replace 'YOUR_SAVE_PATH' with the desired location and filename to save the image
    save_path = './bus.jpg'
    save_path1 = './cat_dog.jpg'
    

    load_and_save_image(image_url, save_path)
    load_and_save_image(image_url1, save_path1)
```
